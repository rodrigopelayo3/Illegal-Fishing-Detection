```python
def parse_event_vessels(event_vessels_str):
    try:
        # Convert the JSON string to a list of dictionaries
        vessels_list = json.loads(event_vessels_str.replace("'", "\""))
        
        # Create lists to store extracted data
        vessel_ids = []
        vessel_types = []
        vessel_ssvids = []
        vessel_names = []
        vessel_flags = []
        vessel_rfmo_authorizations = []
        vessel_origin_ports = []
        vessel_destination_ports = []
        
        # Extract details for each vessel
        for vessel in vessels_list:
            vessel_ids.append(vessel.get("id", None))
            vessel_types.append(vessel.get("type", None))
            vessel_ssvids.append(vessel.get("ssvid", None))
            vessel_names.append(vessel.get("name", None))
            vessel_flags.append(vessel.get("flag", None))
            rfmo_auth = vessel.get("authorizations", [{}])[0].get("rfmo", None)
            vessel_rfmo_authorizations.append(rfmo_auth)
            origin_port_label = vessel.get("origin_port", {}).get("label", None)
            vessel_origin_ports.append(origin_port_label)
            destination_port_label = vessel.get("destination_port", {}).get("label", None)
            vessel_destination_ports.append(destination_port_label)
        
        return pd.Series([vessel_ids, vessel_types, vessel_ssvids, vessel_names, vessel_flags, vessel_rfmo_authorizations, vessel_origin_ports, vessel_destination_ports])
    except (ValueError, TypeError):
        # Handle cases where JSON parsing fails
        return pd.Series([None]*8)
```

```python
# Apply the function to the 'event_vessels' column to create new columns
combined_df[['vessel_ids', 'vessel_types', 'vessel_ssvids', 'vessel_names', 'vessel_flags', 'vessel_rfmo_authorizations', 'vessel_origin_ports', 'vessel_destination_ports']] = combined_df['event_vessels'].apply(parse_event_vessels)

# Explode the lists into separate rows
df_exploded = combined_df.explode(['vessel_ids', 'vessel_types', 'vessel_ssvids', 'vessel_names', 'vessel_flags', 'vessel_rfmo_authorizations', 'vessel_origin_ports', 'vessel_destination_ports'])
```
