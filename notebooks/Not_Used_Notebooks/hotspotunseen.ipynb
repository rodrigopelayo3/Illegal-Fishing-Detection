```python
import pandas as pd

# Load the Welch et al. dataset from a .txt file
df = pd.read_csv('/Users/rodrigo/Desktop/BrainStation/Capstone/DATASETSDIF/HotspotUnseen/hotspot.txt', delimiter=',')
```

```python
df.head()
```

```python
# Inspect the first few rows of the dataframe
print("AIS Disabling Events Dataset:")
print(df.head())

# Check for missing values in the dataframe
print("\nMissing values in AIS Disabling Events Dataset:")
print(df.isnull().sum())

# Check the data types in the dataframe
print("\nData types in AIS Disabling Events Dataset:")
print(df.dtypes)
```

```python
# Handle missing values
df['flag'].fillna('Unknown', inplace=True)
df['vessel_length_m'].fillna(df['vessel_length_m'].median(), inplace=True)

# Convert timestamps to datetime
df['gap_start_timestamp'] = pd.to_datetime(df['gap_start_timestamp'])
df['gap_end_timestamp'] = pd.to_datetime(df['gap_end_timestamp'])

# Check the dataframe after processing
print("\nProcessed AIS Disabling Events Dataset:")
print(df.info())
print(df.head())

# Save the cleaned and processed dataset
df.to_csv('/Users/rodrigo/Desktop/BrainStation/Capstone/DATASETSDIF/processed_ais_disabling_events.csv', index=False)
print("\nData saved to 'processed_ais_disabling_events.csv'")
```

```python
# Optionally, drop rows with missing critical values or fill them with appropriate values
df.dropna(inplace=True)
```

```python
# Standardize column names
df.rename(columns={
    'gap_id': 'event_id',
    'mmsi': 'vessel_id',
    'vessel_class': 'vessel_class',
    'flag': 'flag_state',
    'vessel_length_m': 'vessel_length_m',
    'vessel_tonnage_gt': 'vessel_tonnage_gt',
    'gap_start_timestamp': 'start_timestamp',
    'gap_start_lat': 'start_latitude',
    'gap_start_lon': 'start_longitude',
    'gap_start_distance_from_shore_m': 'start_distance_from_shore_m',
    'gap_end_timestamp': 'end_timestamp',
    'gap_end_lat': 'end_latitude',
    'gap_end_lon': 'end_longitude',
    'gap_end_distance_from_shore_m': 'end_distance_from_shore_m',
    'gap_hours': 'gap_hours'
}, inplace=True)

# Confirm changes
print(df.head())
```

```python
# Convert 'start_timestamp' and 'end_timestamp' to datetime format
df['start_timestamp'] = pd.to_datetime(df['start_timestamp'])
df['end_timestamp'] = pd.to_datetime(df['end_timestamp'])

# Optionally, extract additional time-based features
df['start_year'] = df['start_timestamp'].dt.year
df['start_month'] = df['start_timestamp'].dt.month
df['start_day'] = df['start_timestamp'].dt.day

df['end_year'] = df['end_timestamp'].dt.year
df['end_month'] = df['end_timestamp'].dt.month
df['end_day'] = df['end_timestamp'].dt.day
```

```python
# Save the cleaned data to a CSV file
df.to_csv('/Users/rodrigo/Desktop/BrainStation/Capstone/DATASETSDIF/HotspotUnseen/cleaned_hotspot.csv', index=False)
```
